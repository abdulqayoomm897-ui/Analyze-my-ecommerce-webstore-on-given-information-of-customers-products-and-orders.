```python
import pandas as pd
# Analyze my ecommerce webstore on given information of customers, products and orders.

# Analysis:

# Find how many customers are repeating on my store.
# Find which product is top selling product on the basis of repeated orders.
# Find which product should be discontinued as it is the least soled product on my store.
# Find which customer should get more discount as he/she keep visiting my store more than 5 times.
# What is my total income till end.
# Write me a full report on yearly sale.
# Write me a full report on monthly sale.
```

```python
order = pd.read_csv("orders.csv",na_values=["Fail","Absent"])
product = pd.read_csv("products.csv")
customer = pd.read_csv("customers.csv")
order.head()
```

```python
order.info()
```

```python
order.order_date = pd.to_datetime(order.order_date)
```

```python
 # Find how many customers are repeating on my store.
merged_data = pd.merge(order, customer, on="customer_id", how="left")
order_counts = merged_data["customer_id"].value_counts()
# Filter only those customers who appear more than one time
repeating_customers = order_counts[order_counts > 1]
# Show total number of repeating customers
print("Total repeating customers:", len(repeating_customers))
```

```python

# here we get the top product sold in 2023  --> mtlb 2023 me sb zyyada sell hony wali products
top_products = order[order["order_date"].dt.year ==2023]["product_id"].value_counts
top_products
```

```python
# least sold product that must be eliminated
product_sales = order["product_id"].value_counts()
least_sold_product_id = product_sales.idxmin()
least_sold_product = product[product["product_id"] == least_sold_product_id]

print("Least sold product:",least_sold_product)
```

```python
# Find which customer should get more discount as he/she keep visiting my store more than 5 times.
print("      <<Discount applied for these loyal customers >>!")
customers_visited_moreThan_5_times = order_counts[order_counts > 5]
print("Customers who ordered more than 5 times:",customers_visited_moreThan_5_times)
```

```python
# What is my total income till end.
revenue = product["price"].sum()

print("Total Store Revenue:", revenue)
```
